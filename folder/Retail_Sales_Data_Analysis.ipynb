Handling Null values

```python
import pandas as pd
df=pd.read_csv('orders.csv',na_values=['Not Available','unknown'] )
```

```python
df['Ship Mode'].unique()
```

```python
df.head(30)
```

```python
df = df.rename(columns={'Ship Mode':'Shipping'})
```

```python
df.head(20)
```

handling multiple columns at once

```python
df.columns = df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(" ","_")
```

```python
df.head(5)
```

Deriving New columns

```python
df['Actual_Discount'] = df['list_price']*(df['discount_percent']/100)
```

```python
df['Sale_Price'] = df['list_price'] - df['Actual_Discount']
```

```python
df['Profit'] = df['Sale_Price'] - df['cost_price']
```

```python
df.head(5)
```

Covert date object into date type

```python
df.dtypes
```

```python
df['order_date']=pd.to_datetime(df['order_date'], format="%Y-%M-%d")
```

drop unneccessary columns

```python
df.drop(columns=['list_price','discount_percent','cost_price'],inplace=True)
```

```python
from google.colab import files

df.to_csv('ordersf.csv', index=False)
files.download('ordersf.csv')
```
